```python
#Meslier Passes
mes_pass = []

for i in range(len(event[0])):
    try:
        if event[0][i]['playerId'] == 300447:
            if event[0][i]['type']['displayName'] == 'Pass':
                mes_pass.append(event[0][i])
    except:
        pass
```

```python
mes_pass[0]
```

```python
mes_succ_pass_map = []
mes_unsucc_pass_map = []

for i in range(len(mes_pass)):
    each_pass = []
    each_pass.append(mes_pass[i]['x'])
    each_pass.append(mes_pass[i]['y'])
    each_pass.append(mes_pass[i]['endX'])
    each_pass.append(mes_pass[i]['endY'])
    each_pass.append(mes_pass[i]['outcomeType']['displayName'])
    if mes_pass[i]['outcomeType']['displayName'] == 'Successful':
        mes_succ_pass_map.append(each_pass)
    else:
        mes_unsucc_pass_map.append(each_pass)
```

```python
df_succ_pass = pd.DataFrame(mes_succ_pass_map, columns = ['StartX','StartY','EndX','EndY','Outcome'])
df_unsucc_pass = pd.DataFrame(mes_unsucc_pass_map, columns = ['StartX','StartY','EndX','EndY','Outcome'])
```

```python
df_succ_pass
```

```python
# Setup the pitch
pitch = Pitch(pitch_type='opta', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=False, tight_layout=True)
fig.set_facecolor('#22312b')

# Plot the completed passes
lc1 = pitch.lines(df_succ_pass.StartX, df_succ_pass.StartY,
                  df_succ_pass.EndX, df_succ_pass.EndY,
                  lw=5, transparent=True, comet=True, label='completed passes',
                  color='#ad993c', ax=ax)

# Plot the other passes
lc2 = pitch.lines(df_unsucc_pass.StartX, df_unsucc_pass.StartY,
                  df_unsucc_pass.EndX, df_unsucc_pass.EndY,
                  lw=5, transparent=True, comet=True, label='other passes',
                  color='#ba4f45', ax=ax)

# Plot the legend
ax.legend(facecolor='#22312b', edgecolor='None', fontsize=20, loc='upper left', handlelength=4)
```
