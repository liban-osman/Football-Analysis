### Importing Dependencies

```python
import pandas as pd
import numpy as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

from functions import players_list, load_data
from mplsoccer import Pitch, VerticalPitch, FontManager
```

### Finding match specific data

```python
match = "matches/liv_manu.json"
data = load_data(match)
```

### Initializing Player and Event Data

```python
players = players_list(data)
events = pd.DataFrame(data["events"])
```

```python
players
```

### Pass Analysis

```python
events
```

```python
passes = events[events["type"] == {'value': 1, 'displayName': 'Pass'}]
```

```python
passes.columns
```

```python
pass_degea = passes[passes["playerId"] == 79554]
```

```python
df_succ_pass = passes[passes["outcomeType"] == {'value': 1, 'displayName': 'Successful'}]
df_unsucc_pass = passes[passes["outcomeType"] == {'value': 0, 'displayName': 'Unsuccessful'}]
```

```python
# Setup the pitch
pitch = Pitch(pitch_type='opta', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=False, tight_layout=True)
fig.set_facecolor('#22312b')

# Plot the completed passes
lc1 = pitch.lines(df_succ_pass.x, df_succ_pass.y,
                  df_succ_pass.endX, df_succ_pass.endY,
                  lw=5, transparent=True, comet=True, label='completed passes',
                  color='#ad993c', ax=ax)

# Plot the other passes
lc2 = pitch.lines(df_unsucc_pass.x, df_unsucc_pass.y,
                  df_unsucc_pass.endX, df_unsucc_pass.endY,
                  lw=5, transparent=True, comet=True, label='other passes',
                  color='#ba4f45', ax=ax)

# Plot the legend
ax.legend(facecolor='#22312b', edgecolor='None', fontsize=20, loc='upper left', handlelength=4)
```
